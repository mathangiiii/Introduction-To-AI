```python

#graph visualisation
import networkx as nx
nodes = ['S','A','B','C','G']
#Depth Limited Search
graph = {
    'a': ['b','c'],
    'b': ['d','e'],
    'c': ['f','g'],
    'd': [],
    'e': ['h','i'],
    'f': [],
    'g': [],
    'h': [],
    'i': []}
G=nx.Graph()
for key in graph.keys():
    for i in range(len(graph[key])):
        G.add_edge(key,graph[key][i])
nx.draw(G,with_labels=True)                   
```

```python
#Iterative Deepening Search
#remove explored parameter and make it global
explored = set()
def dls1( graph, crnt_state,goal_state,limit):
    global explored
    if crnt_state == goal_state:
        print(crnt_state,end=" ")
        return True
    if crnt_state not in explored:
        print(crnt_state,end=" ")
        explored.add(crnt_state)
        if limit >= 0:
            found_goal = False
            for adj in graph[crnt_state]:
                if dls1(graph,adj,goal_state,limit-1):
                    found_goal = True
                    break
            return found_goal        
    return False        
      
def ids(graph,initial_state,goal_state,max_depth):
    global exlored
    print('IDS:')
    for limit in range(max_depth+1):
        explored.clear()
        print("\ndepth ",limit,':')
        if dls1( graph, initial_state,goal_state,limit):
            print("\nGoal state reached at depth",limit)
            return
ids(graph,'a','g',2)        
```
