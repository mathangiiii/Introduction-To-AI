```python
#graph visualisation
import networkx as nx
nodes = ['S','A','B','C','G']
graph = {
    'S':['A','B','C'],
    'A':['S','G'] ,
    'B':['S','G'] ,
    'C':['S','G'] ,
    'G':['A','B','C']
}
G=nx.Graph()
for key in graph.keys():
    for i in range(len(graph[key])):
        G.add_edge(key,graph[key][i])
nx.draw(G,with_labels=True)                   
```

```python
#bfs
def bfs(graph,initial):
    V=len(graph)
    visited = {
    'S':False ,
    'A':False ,
    'B':False ,
    'C':False ,
    'G':False
}
    frontier = []
    path = []
    frontier.append(initial)
    visited[initial] = True
    while frontier:
        node = frontier.pop(0)
        path.append(node)
        for i in graph[node]:
            if visited[i] == False:
                frontier.append(i)
                visited[i] = True
    print('->'.join(path))                    
bfs(graph,'S')               
```
